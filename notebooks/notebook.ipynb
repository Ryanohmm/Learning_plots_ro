```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

```python
yearly_avg = spotify_df.groupby("Release Year")["Total Streams (Millions)"].mean().round(2)
yearly_avg
```

```python
plt.figure(figsize=(10,6))
plt.plot(
    yearly_avg.index, 
    yearly_avg.values,
    marker="+",
    linestyle = "--",
    color = "#05830f",
    linewidth = 2)

plt.xlabel("Release Year")
plt.ylabel("Average Total Streams")
plt.title("Did Listeners Stream Less in 2022?", fontsize = 18, pad=20)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)

ax.axvline(
    2022,
    color="green",
    linewidth = 0.5,
    alpha = 0.7)

    ax.text(
        2022+0.05,
        3000,
        "Dip In Total Streams"
        color="green"
    )

plt.ylim(0, 3500)

plt.show
```

```python
category = "Platform Type"
values = "Total Streams (Millions)"

platform_totals = spotify_df.groupby(category)[values].sum()
platform_totals
```

```python
plt.Figure(figsize=(8,8))
plt.pie(
    platform_totals,
    labels=None, 
    startangle=90,
    colors=["#grey", "#08881d"],
    autopct="%1.1f%%")
plt.title("Near Even SPlit Between Premium and Free Users")

plt.legend(
    labels = platform_totals.index,
    loc="lower right",
    title = "Platform"
)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

wedges, texts, autotexts = ax.pie(
    platform_totals, 
    autopct="%1.1f%%",
    startangle=90,
    colors=["#grey", "#08881d"])

    ax.set_title(
        "Near Even Split Between Premium and Free Users"
        fontsize=18,
        pad=15,
        loc="center"
    )
for autotext in autotext:
autotext.set_fontsize(25)
autotext.set_color("white")
autotext.set_weight("bold")


ax.legend(
    wedges,
    labels = platform_totals.index,
    loc="under center",
    bbox_to_anchor = [0.5, 0.95],
    ncol=2
)

plt.tight_layout()

plt.show()
```

```python
plt.Figure(figsize=(8,8))

plt.hist(
    spotify_df["Avg Stream Duration (Min)"],
    bins=20,
    color="#08881d",
    edgecolor = "white"
)

plt.xlabel("Average Stream Duration (Minutes)")
plt.ylabel("Number of Tracks")
plt.title("Distribution of Average Stream Duration", fontsize=16, pad=20)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig("../plots/ploticuswrecks.png")
plt.show()
```
